```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')  # or pd.read_csv('titanic.csv')

# Quick look
titanic.head()
titanic.info()
titanic.describe(include='all')
```

```python
titanic.isnull().sum()
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
```

```python
# Numerical variables
titanic['age'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')

# Categorical variables
sns.countplot(x='sex', data=titanic, palette='Set2')
plt.title('Count of Passengers by Sex')
```

```python
sns.boxplot(x='sex', y='age', data=titanic, palette='pastel')
sns.barplot(x='class', y='survived', data=titanic, palette='coolwarm')
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
```

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='husl')
```
